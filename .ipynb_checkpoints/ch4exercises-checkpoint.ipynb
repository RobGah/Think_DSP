```python
"""Exercise 4.2"""

#implement Bartlett's Method for spectral power estimation
#based on author's make_spectrogram
#this was a KILLER...had to look at the answer.

def bartlett_method(wave, seg_length=512, win_flag=True):
    """Estimates the power spectrum of a noise wave.
    
    wave: Wave
    seg_length: segment length
    """
    # make a spectrogram and extract the spectrums
    spectro = wave.make_spectrogram(seg_length, win_flag) 
   # author used make_spectrogram which has a 1/2 segment overlap...Bartletts should have no overlap?
    spectrums = spectro.spec_map.values() #grabs the spec_map dictionary from the spectrum object
   
    # extract the power array from each spectrum
    psds = [spectrum.power for spectrum in spectrums] #array of psd arrays for each segment's spectrum 
    
    # compute the root mean power (which is like an amplitude)
    hs = np.sqrt(sum(psds) / len(psds)) #like an rms...makes sense
    fs = next(iter(spectrums)).fs #this was also something I struggled with - how to get the fs array for each segment 
    
    # make a Spectrum with the mean amplitudes
    spectrum = thinkdsp.Spectrum(hs, fs, wave.framerate)
    return spectrum

#in any event, I learned something...and I know what's going on here now. 

#I'll do what the author did and create PSDs for both my original segment 
#and a new segment from the crowd noise and compare
crowd_seg2 = crowd.segment(start = 30 , duration = 20)

psd = bartlett_method(crowd_seg)
psd2 = bartlett_method(crowd_seg2)

psd.plot_power()
psd2.plot_power(color = 'red')
plt.xscale('log')
plt.yscale('log')

```

```python
"""
Exercise 4.3
"""
import pandas as pd
import matplotlib.pyplot as plt
daily_BTC = pd.read_csv('BTC_USD_2019-03-01_2019-05-31-CoinDesk.csv',parse_dates = True, usecols=['Date','Closing Price (USD)'])
daily_BTC.info()
hs = daily_BTC['Closing Price (USD)'].get_values()
ts = np.linspace(1,len(hs),len(hs))
#print(ts) #a check that we have the right ts
plt.plot(ts,hs) #looks the same as what coindesk puts out
plt.xlabel('days')
plt.ylabel('USD')
```

```python
wave = thinkdsp.Wave(hs,ts)
spectrum = wave.make_spectrum()
spectrum.plot_power()
plt.xscale('log')
plt.yscale('log')
plt.ylabel('power')
plt.xlabel('frequency')
spectrum.estimate_slope()[0] #get slope estimate
#I tried a few different lengths of time. The greater the length of time...the more it looks like red noise
#24 hours = -0.93
#3months = -1.968
#6months = -1.874 ...meh...maybe this one wasnt as much of a random walk...some big crypto event happen??
#1year = -1.993
```

```python
"""
Exercise 4.4
"""
class UncorrelatedPoissonNoise(thinkdsp._Noise):
    def evaluate(self,ts):
        ys = np.random.poisson(self.amp,len(ts))
        return ys
frame = 10000 #use for comparison
time  = 1 
poisson = UncorrelatedPoissonNoise(amp=.01)
wave_poisson = poisson.make_wave(duration = time, framerate = frame)
display(wave_poisson.make_audio()) #sounds like a geiger counter at low lam (expected arrival rate)
poisson_spectrum = wave_poisson.make_spectrum()
white = thinkdsp.UncorrelatedGaussianNoise(amp=1)
white_wave = white.make_wave(duration = time, framerate = frame)
white_spectrum = white_wave.make_spectrum()
white_spectrum.plot_power(alpha=0.7,color='orange',label='white')
poisson_spectrum.plot_power(alpha=0.7,label='poisson')
plt.legend(loc='lower left')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('freq')
plt.ylabel('power')
```

```python
white_spectrum.plot(alpha=0.7, color='orange',label='white')
poisson_spectrum.plot(alpha=0.7, label='poisson')
plt.xlabel('freq')
plt.ylabel('amplitude')
plt.legend(loc='upper right')
```
